```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
# Read in the CSV files
df1 = pd.read_csv("/Users/vp/Desktop/MDSE/2S/EGD/project/EGD_Project/performance-csv/2-cores-6-instances.csv")
df2 = pd.read_csv("/Users/vp/Desktop/MDSE/2S/EGD/project/EGD_Project/performance-csv/2-cores-12-instances.csv")
df3 = pd.read_csv("/Users/vp/Desktop/MDSE/2S/EGD/project/EGD_Project/performance-csv/2-cores-18-instances.csv")

# Concatenate the dataframes
df = pd.concat([df1, df2, df3], ignore_index=True)
df['Query/Model'] = df['Query/Model'].replace('Liner Regression', 'Linear Regression')
```

```python
# 2 cores & 6 instances
# Create subplots
fig, axs = plt.subplots(nrows=3, ncols=3, figsize=(40, 15))
data1 = df[(df['Cores & Instances'] == '2 cores & 6 instances') & (df['Dataset size'] == 100)].reset_index(drop=True)
data1['Model_label'] = pd.factorize(data1['Query/Model'])[0]
data2 = df[(df['Cores & Instances'] == '2 cores & 6 instances') & (df['Dataset size'] == 700)].reset_index(drop=True)
data2['Model_label'] = pd.factorize(data2['Query/Model'])[0]
data3 = df[(df['Cores & Instances'] == '2 cores & 6 instances') & (df['Dataset size'] == 1300)].reset_index(drop=True)
data3['Model_label'] = pd.factorize(data3['Query/Model'])[0]
# Column 1
axs[0, 0].bar(data1['Model_label'], data1['Execution Time'],width=0.25, label='Execution Time')
axs[0, 0].bar(data1['Model_label']+0.25, data1['CPU Usage'],width=0.25, label='CPU Usage')
axs[0, 0].set_xticks([])
axs[0, 0].set_title(r'$\bf{2\ cores\ and\ 6\ instances}$' +'\n\nDataset size: 100 MB',fontsize=15)
axs[0, 0].legend()
# Column 2
axs[1, 0].bar(data2['Model_label'], data2['Execution Time'],width=0.25, label='Execution Time')
axs[1, 0].bar(data2['Model_label']+0.25, data2['CPU Usage'],width=0.25, label='CPU Usage')
axs[1, 0].set_xticks([])
axs[1, 0].set_title('\nDataset size: 700 MB', fontsize=15)
axs[1, 0].legend()
# Coumn 3
axs[2, 0].bar(data3['Model_label'], data3['Execution Time'],width=0.25, label='Execution Time')
axs[2, 0].bar(data3['Model_label']+0.25, data3['CPU Usage'],width=0.25, label='CPU Usage')
axs[2, 0].set_xticks(range(len(data3)))
axs[2, 0].set_xticklabels(data3['Query/Model'].values, rotation=45)
axs[2, 0].legend()
# 2 cores & 12 instances
data1 = df[(df['Cores & Instances'] == '2 cores & 12 instances') & (df['Dataset size'] == 100)].reset_index(drop=True)
data1['Model_label'] = pd.factorize(data1['Query/Model'])[0]
data2 = df[(df['Cores & Instances'] == '2 cores & 12 instances') & (df['Dataset size'] == 700)].reset_index(drop=True)
data2['Model_label'] = pd.factorize(data2['Query/Model'])[0]
data3 = df[(df['Cores & Instances'] == '2 cores & 12 instances') & (df['Dataset size'] == 1300)].reset_index(drop=True)
data3['Model_label'] = pd.factorize(data3['Query/Model'])[0]
# Column 1
axs[0, 1].bar(data1['Model_label'], data1['Execution Time'],width=0.25, label='Execution Time')
axs[0, 1].bar(data1['Model_label']+0.25, data1['CPU Usage'],width=0.25, label='CPU Usage')
axs[0, 1].set_xticks([])
axs[0, 1].set_title(r'$\bf{2\ cores\ and\ 12\ instances}$' + '\n\nDataset size: 100 MB',fontsize=15)
axs[0, 1].legend()
# Column 2
axs[1, 1].bar(data2['Model_label'], data2['Execution Time'],width=0.25, label='Execution Time')
axs[1, 1].bar(data2['Model_label']+0.25, data2['CPU Usage'],width=0.25, label='CPU Usage')
axs[1, 1].set_xticks([])
axs[1, 1].set_title('\nDataset size: 700 MB', fontsize=15)
axs[1, 1].legend()
# Column 3
axs[2, 1].bar(data3['Model_label'], data3['Execution Time'],width=0.25, label='Execution Time')
axs[2, 1].bar(data3['Model_label']+0.25, data3['CPU Usage'],width=0.25, label='CPU Usage')
axs[2, 1].set_xticks(range(len(data3)))
axs[2, 1].set_xticklabels(data3['Query/Model'].values, rotation=45)
axs[2, 1].set_title('\nDataset size: 1300 MB',fontsize=15)
axs[2, 1].legend()
# 2 cores & 18 instances
data1 = df[(df['Cores & Instances'] == '2 cores & 18 instances') & (df['Dataset size'] == 100)].reset_index(drop=True)
data1['Model_label'] = pd.factorize(data1['Query/Model'])[0]
data2 = df[(df['Cores & Instances'] == '2 cores & 18 instances') & (df['Dataset size'] == 700)].reset_index(drop=True)
data2['Model_label'] = pd.factorize(data2['Query/Model'])[0]
data3 = df[(df['Cores & Instances'] == '2 cores & 18 instances') & (df['Dataset size'] == 1300)].reset_index(drop=True)
data3['Model_label'] = pd.factorize(data3['Query/Model'])[0]
# Column 1
axs[0, 2].bar(data1['Model_label'], data1['Execution Time'],width=0.25, label='Execution Time')
axs[0, 2].bar(data1['Model_label']+0.25, data1['CPU Usage'],width=0.25, label='CPU Usage')
axs[0, 2].set_xticks(range(len(data1)))
axs[0, 2].set_xticklabels(data1['Query/Model'].values, rotation=45)
axs[0, 2].set_title(r'$\bf{2\ cores\ and\ 18\ instances}$' + '\n\nDataset size: 100 MB',fontsize=15)
axs[0, 2].legend()
# Column 2
axs[1, 2].bar(data2['Model_label'], data2['Execution Time'],width=0.25, label='Execution Time')
axs[1, 2].bar(data2['Model_label']+0.25, data2['CPU Usage'],width=0.25, label='CPU Usage')
axs[1, 2].set_xticks(range(len(data2)))
axs[1, 2].set_xticklabels(data2['Query/Model'].values, rotation=45)
axs[1, 2].set_title('\nDataset size: 700 MB', fontsize=15)
axs[1, 2].legend()
# Column 3
axs[2, 2].bar(data3['Model_label'], data3['Execution Time'],width=0.25, label='Execution Time')
axs[2, 2].bar(data3['Model_label']+0.25, data3['CPU Usage'],width=0.25, label='CPU Usage')
axs[2, 2].set_xticks(range(len(data3)))
axs[2, 2].set_xticklabels(data3['Query/Model'].values, rotation=45)
axs[2, 2].set_title('\nDataset size: 1300 MB')
axs[2, 2].legend()
fig.suptitle('Performance metrics for different Cores & Instances configurations', fontsize=25,fontweight="bold")
for ax in axs.flat:
     ax.label_outer()
#fig.tight_layout()
```
